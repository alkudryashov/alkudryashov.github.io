```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
f = lambda x, alpha=0, beta=0: (x**beta + alpha**beta)/x**beta
```

```python
import sys
sys.version_info
```

# Задание 1. Часть 1

Построить графики в общих осях для:
1. $\alpha = 1, \beta = 1$
2. $\alpha = 2, \beta = 1$
3. $\alpha = 1, \beta = 2$

```python
x = np.linspace(-5, 5, 501)
f1 = f(x, 1, 1)
f2 = f(x, 2, 1)
f3 = f(x, 1, 2)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, f1, "r")
axes.plot(x, f2, "g")
axes.plot(x, f3, "b")
axes.set_xlabel("x")
axes.set_ylabel("f(x)")
axes.set_ylim([-10, 10])
axes.set_title(r"График функции $ f(x) = \frac{x ^ \beta + \alpha ^ \beta}{x ^ \beta}$")
axes.legend([r" $\alpha = 1, \beta = 1$", r"$\alpha = 2, \beta = 1$", r"$\alpha = 1, \beta = 2$"]);
```

```python
fig.savefig("1.1.svg")
```

# Задание 1. Часть 2

1. Построить в общих осях графики для x>0.
2. На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:
    для малых x
    для больших x
3. Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений α и β.

```python
x = np.linspace(0.5, 5, 200)
f1 = f(x, 1, 1)
f2 = f(x, 2, 1)
f3 = f(x, 1, 2)

x_small = np.linspace(0.1, 0.5, 200)
f1_small = f(x_small, 1, 1)
f2_small = f(x_small, 2, 1)
f3_small = f(x_small, 1, 2)

x_big = np.linspace(100, 1000, 200)
f1_big = f(x_big, 1, 1)
f2_big = f(x_big, 2, 1)
f3_big = f(x_big, 1, 2)

fig = plt.figure(figsize=(13, 8))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f1, "r")
axes1.plot(x, f2, "g")
axes1.plot(x, f3, "b")
axes1.set_xlabel("x", fontsize=18)
axes1.set_ylabel("f(x)", fontsize=18)
axes1.set_title(r"График функции $ f(x) = \frac{x ^ \beta + \alpha ^ \beta}{x ^ \beta}, x > 0 $", fontsize=18)
axes1.legend([r" $\alpha = 1, \beta = 1$", r"$\alpha = 2, \beta = 1$", r"$\alpha = 1, \beta = 2$"], loc=5)

axes2 = fig.add_axes([0.4, 0.7, 0.2, 0.2])
axes2.plot(x_small, f1_small, "r")
axes2.plot(x_small, f2_small, "g")
axes2.plot(x_small, f3_small, "b")
axes2.set_title("Для малых x")

axes3 = fig.add_axes([0.7, 0.7, 0.2, 0.2])
axes3.plot(x_big, f1_big, "r")
axes3.plot(x_big, f2_big, "g")
axes3.plot(x_big, f3_big, "b")
axes3.set_title("Для больших x");
```

```python
fig.savefig("1.2.svg")
```

# Задание 1. Часть 3

1. Построить в общих осях графики для x<0.
2. На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к −∞.
3. Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. 
4. Так же нанесите на графики прямую f(x) = 0.Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений α и β.

```python
x = np.linspace(-5, -0.5, 200)
f1 = f(x, 1, 1)
f2 = f(x, 2, 1)
f3 = f(x, 1, 2)

x_small = np.linspace(-1000, -20, 200)
f1_small = f(x_small, 1, 1)
f2_small = f(x_small, 2, 1)
f3_small = f(x_small, 1, 2)

fig = plt.figure(figsize=(13, 8))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f1, "r")
axes1.plot(x, f2, "g")
axes1.plot(x, f3, "b")
axes1.axhline(0, 0.05, 0.95, color="black", linestyle="--")
axes1.set_xlabel("x", fontsize=18)
axes1.set_ylabel("f(x)", fontsize=18)
axes1.set_title(r"График функции $ f(x) = \frac{x ^ \beta + \alpha ^ \beta}{x ^ \beta}, x < 0 $", fontsize=18)
axes1.legend([r" $\alpha = 1, \beta = 1$", r"$\alpha = 2, \beta = 1$", r"$\alpha = 1, \beta = 2$", "y=0"], loc=5)

axes2 = fig.add_axes([0.2, 0.6, 0.5, 0.3])
axes2.plot(x_small, f1_small, "r")
axes2.plot(x_small, f2_small, "g")
axes2.plot(x_small, f3_small, "b")
#axes2.axhline(0, 0.05, 0.95, color="black", linestyle="--")
axes2.set_title(r"При $x \rightarrow -\inf$");
```

```python
fig.savefig("1.3.svg")
```

## Выводы: при x < 0 функции не имеют общих точек и  не пересекаются, хоть и стремятся к ассимптоте $y=1$, каждая со своей стороны

# Задача 3

Построить в общих осях графики
1. $\alpha = 1, \beta = 0.5$
2. $\alpha = 1, \beta = -0.5$
6. $\alpha = 1, \beta = -1.5$

Сделать выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
xgz = np.linspace(0.1, 5, 201)
xlz = np.linspace(-5, -0.1, 201)
f1gz = f(xgz, 1, 0.5)
f2gz = f(xgz, 1, -0.5)
f3gz = f(xgz, 1, -1.5)
f1lz = f(xlz, 1, 0.5)
f2lz = f(xlz, 1, -0.5)
f3lz = f(xlz, 1, -1.5)
```

```python
np.isnan(f1lz).all(), np.isnan(f2lz).all(), np.isnan(f3lz).all()
```

### Заметим, что для всех x < 0 функции с такими параметрами не определены

```python
fig = plt.figure(figsize=(13, 8))
plt.plot(xgz, f1gz, "r")
plt.plot(xgz, f2gz, "g")
plt.plot(xgz, f3gz, "b")
plt.xlabel("x", fontsize=18)
plt.ylabel("f(x)", fontsize=18)
plt.title(r"График функции $ f(x) = \frac{x ^ \beta + \alpha ^ \beta}{x ^ \beta}$ при  $x>0$", fontsize=14)
plt.legend([r"$\alpha = 1, \beta = 0.5$",
              r"$\alpha = 1, \beta = -0.5$",
              r"$\alpha = 1, \beta = -1.5$"], loc=5);
```

```python
fig.savefig("3.svg")
```

## Выводы: функции пересекаются в точке (1, 2), они определены при x >0, при этом функция с параметрами $\alpha=1, \beta=-1.5$ всюду возрастает и выпукла, функция с параметрами $\alpha=1, \beta=0.5$ всюду убывает и выпукла, функция с параметрами $\alpha=1, \beta=-0.5$ всюду возврастает и вогнута

# Задача 4

Постройте в одном ряду 3 графика, чтобы убедиться в выводах, сделанных по результатам предыдущей задачи.

Каждый график будет содержать 4 кривые. 2 общих:
1. $\alpha = 1, \beta = 0$
2. $\alpha = 1, \beta = -1$

И по два уникальных графика: 
1. $\alpha = 1, \beta = 0.5$ и $\alpha = 1, \beta = 0.8$
2. $\alpha = 1, \beta = -0.5$ и $\alpha = 1, \beta = -0.8$
6. $\alpha = 1, \beta = -1.5$ и $\alpha = 1, \beta = -2.5$

Сделать выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x = np.linspace(-5, 5, 203)
f11 = f(x, 1, 0)
f12 = f(x, 1, -1)
f211 = f(x, 1, 0.5)
f212 = f(x, 1, 0.8)
f221 = f(x, 1, -0.5)
f222 = f(x, 1, -0.8)
f231 = f(x, 1, -1.5)
f232 = f(x, 1, -2.5)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 8))

axes[0].plot(x, f11, "b--")
axes[0].plot(x, f12, "r--")
axes[0].set_xlabel("x", fontsize=18)
axes[0].set_ylabel("f(x)", fontsize=18)
axes[0].plot(x, f211, x, f212)
axes[0].legend([r"$\alpha = 1, \beta = 0.5$",
              r"$\alpha = 1, \beta = -0.5$", 
               r"$\alpha = 1, \beta = 0.5$", r"$\alpha=1, \beta=0.8$"])


axes[1].plot(x, f11, "b--")
axes[1].plot(x, f12, "r--")
axes[1].set_xlabel("x", fontsize=18)
axes[1].set_ylabel("f(x)", fontsize=18)
axes[1].plot(x, f221, x, f222)
axes[1].legend([r"$\alpha = 1, \beta = 0.5$",
              r"$\alpha = 1, \beta = -0.5$", 
               r"$\alpha = 1, \beta = -0.5$", r"$\alpha=1, \beta=-0.8$"])

axes[2].plot(x, f11, "b--")
axes[2].plot(x, f12, "r--")
axes[2].set_xlabel("x", fontsize=18)
axes[2].set_ylabel("f(x)", fontsize=18)
axes[2].plot(x, f231, x, f232)
axes[2].legend([r"$\alpha = 1, \beta = 0.5$",
              r"$\alpha = 1, \beta = -0.5$", 
               r"$\alpha = 1, \beta = -1.5$", r"$\alpha=1, \beta=-2.5$"])

plt.suptitle(r"График функции $ f(x) = \frac{x ^ \beta + \alpha ^ \beta}{x ^ \beta}$", fontsize=14);
```
